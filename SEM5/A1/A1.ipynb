### Data Visualization-Create a histogram for each feature in the dataset to illustrate the feature distributions. Plot each histogram.

```python
import matplotlib.pyplot as plt
import seaborn as sns



fig,axes = plt.subplots(2,2,figsize=(16,10))
sns.histplot(data=df,x='SepalLengthCm',ax=axes[0][0],bins=20,kde=True,hue='Species',alpha=0.5)
sns.histplot(data=df,x='SepalWidthCm',ax=axes[0][1],bins=20,kde=True,hue='Species',alpha=0.5)
sns.histplot(data=df,x='PetalLengthCm',ax=axes[1][0],bins=20,kde=True,hue='Species',alpha=0.5)
sns.histplot(data=df,x='PetalWidthCm',ax=axes[1][1],bins=20,kde=True,hue='Species',alpha=0.5)
```

```python
fig,axes = plt.subplots(2,2,figsize=(14,16))
sns.boxplot(data=df,y='SepalLengthCm',x='Species',ax=axes[0][0])
sns.boxplot(data=df,y='SepalWidthCm',x='Species',ax=axes[0][1])
sns.boxplot(data=df,y='PetalLengthCm',x='Species',ax=axes[1][0])
sns.boxplot(data=df,y='PetalWidthCm',x='Species',ax=axes[1][1])
```
